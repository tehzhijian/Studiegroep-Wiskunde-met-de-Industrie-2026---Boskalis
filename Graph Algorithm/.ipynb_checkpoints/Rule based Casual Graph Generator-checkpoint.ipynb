```python
import os
import json
import logging
import networkx as nx
import transformers
from builder import CausalGraphBuilder
```

```python
# 1. FIX: Suppress the "UNEXPECTED" model loading warning
# This tells the transformers library to ignore non-critical weight mismatches
transformers.logging.set_verbosity_error()

# 2. Setup standard logging for your application
logging.basicConfig(level=logging.INFO, format='%(asctime)s - %(levelname)s - %(message)s')
logger = logging.getLogger(__name__)

# 3. Load the Input Data
input_file = 'wiki_math_knowledge_base_api.json'
documents = []
```

```python
try:
    with open(input_file, 'r', encoding='utf-8') as f:
        data = json.load(f)
        # Extract 'raw_text' from the JSON structure
        documents = [item.get('raw_text', '') for item in data if item.get('raw_text')]
    
    logger.info(f"Loaded {len(documents)} text chunks from {input_file}")

except FileNotFoundError:
    logger.error(f"File {input_file} not found. Please check the path.")
    documents = []

# 4. Initialize Builder with Rule-Based Extraction
# Note: You will still see the "Unauthenticated requests" warning unless you set os.environ["HF_TOKEN"]
# This is harmless and just means the download might be slightly slower.
builder = CausalGraphBuilder(
    model_name="mixedbread-ai/mxbai-embed-large-v1",
    extractor_method="rule", 
    confidence_threshold=0.6,
    normalize_nodes=True
)
```

```python
if documents:
    # 4. Index Documents
    logger.info("Building graph...")
    builder.index_documents(documents, batch_size=50, show_progress=True)

    # 5. Save for Reuse (The Important Part)
    
    # --- Format A: Native Builder JSON (Best for Python Reuse) ---
    # This preserves the builder's internal mapping of text variants to node IDs.
    # You can reload this later using builder.load("causal_graph.json")
    builder.save("causal_math_graph_state.json")
    logger.info("Saved reusable state to 'causal_math_graph_state.json'")

    # --- Format B: GraphML (Best for External Analysis) ---
    # This standard format can be opened in Gephi, Cytoscape, or other graph tools.
    G = builder.get_graph()
    nx.write_graphml(G, "math_causal_graph.graphml")
    logger.info("Saved network structure to 'math_causal_graph.graphml'")
    
    # 6. (Optional) Quick Statistics check
    print(f"Graph constructed with {G.number_of_nodes()} nodes and {G.number_of_edges()} edges.")

else:
    logger.warning("No documents to process.")
```

```python
# This will reconstruct the graph structure and re-generate embeddings for the nodes
success = builder.load("causal_graph_state.json")
```

```python
if success:
    print("Graph loaded successfully!")
    
    # 3. Verify it loaded by printing some stats
    stats = builder.get_extraction_statistics()
    print(f"Nodes: {stats['graph_statistics']['nodes']}")
    print(f"Edges: {stats['graph_statistics']['edges']}")
    
    # 4. (Optional) Visualize it again to make sure it looks right
    # builder.visualize_graph("reloaded_graph.html")
else:
    print("Failed to load graph.")
```

```python
# Export to GraphML format
G = builder.get_graph()
nx.write_graphml(G, "causal_graph.graphml")
print("Exported to causal_graph.graphml")
```

```python
import matplotlib.pyplot as plt

# Get the graph object
G = builder.get_graph()

# Setup the plot
plt.figure(figsize=(12, 20))
pos = nx.spring_layout(G, k=0.5)  # k regulates the distance between nodes

# Draw the graph
nx.draw(G, pos, 
        with_labels=True, 
        node_color='lightblue', 
        node_size=500, 
        font_size=8, 
        arrows=True)

plt.title("Causal Graph Visualization")
plt.show()
```
